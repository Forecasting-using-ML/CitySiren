```python
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None     # default='warn'
```

```python
raw_fire = pd.read_csv('./raw/Fire_Incident_Dispatch_Data.csv')
raw_fire.head()
```

```python
# filter useful columns
fire_cols = ['STARFIRE_INCIDENT_ID', 'INCIDENT_DATETIME', 'ALARM_BOX_LOCATION', 'INCIDENT_CLASSIFICATION_GROUP']
data_fire = raw_fire[fire_cols]
data_fire.head(10)
```

```python
# drop NA
print(data_fire.shape[0])
data_fire.dropna(subset = fire_cols, inplace = True)
print(data_fire.shape[0])
```

# filter year 2015

```python
data_fire['timestamp'] = data_fire.apply(lambda L: pd.to_datetime(L['INCIDENT_DATETIME'], format = '%m/%d/%Y %I:%M:%S %p'), axis = 1)
data_fire15 = data_fire[(data_fire['timestamp'] >= '2015-01-01 00:00:00') & (data_fire['timestamp'] < '2016-01-01 00:00:00')]
print('#rows:', data_fire15.shape[0])
data_fire15.head(10)
```

```python
data_fire15.drop(['INCIDENT_DATETIME'], axis = 1, inplace = True)
print('#rows:', data_fire15.shape[0])
data_fire15.head()
```

```python
# check categories
data_fire15['INCIDENT_CLASSIFICATION_GROUP'].unique()
```

```python
# separate medical/non-medical
data_medical = data_fire15[(data_fire15['INCIDENT_CLASSIFICATION_GROUP'] == 'Medical MFAs') | (data_fire15['INCIDENT_CLASSIFICATION_GROUP'] == 'Medical Emergencies')]
data_nonmed = data_fire15[(data_fire15['INCIDENT_CLASSIFICATION_GROUP'] != 'Medical MFAs') & (data_fire15['INCIDENT_CLASSIFICATION_GROUP'] != 'Medical Emergencies')]
print('ems #rows:', data_medical.shape[0])
print('fire #rows:', data_nonmed.shape[0])
```

```python
# EMS
data_medical.head()
```

```python
# fire
data_nonmed.head()
```

# Geocoding w/ Google Maps API

```python
from geopy.geocoders import GoogleV3     # pip install geopy
from geopy.extra.rate_limiter import RateLimiter
```

```python
# check unique box locations
box_loc = pd.DataFrame(data_fire15['ALARM_BOX_LOCATION'].drop_duplicates())
print(box_loc.shape[0])
box_loc.head(10)
```

```python
# add city name for geocoding
def addNYC(adr):
    return str(adr) + ', NYC'

box_loc['address'] = box_loc['ALARM_BOX_LOCATION'].apply(addNYC)
box_loc.head(10)
```

```python
# geocoder
geolocator = GoogleV3('YOUR-GCP-CREDENTIAL')
geocode = RateLimiter(geolocator.geocode, min_delay_seconds = 1)
box_loc['location'] = box_loc['address'].apply(geocode)
```

```python
# fetch Latitude/Longitude
def lst2loc(loc_lst):
    if loc_lst is not None:
        lat = loc_lst.latitude
        lon = loc_lst.longitude
    else:
        lat = None
        lon = None
    return lat, lon

box_loc['lat'], box_loc['lon'] = zip(*box_loc['location'].map(lst2loc))
box_loc.head(10)
```

```python
# save geocoding result
box_loc.to_csv('./raw/box_location_raw.csv')
```

# join & save

```python
# load geocoded result
box_loc = pd.read_csv('./raw/box_location_raw.csv')
box_loc.head()
```

```python
# EMS
medical_join = pd.merge(data_medical, box_loc, on = 'ALARM_BOX_LOCATION')
print(medical_join.shape[0])
medical_join.dropna(inplace = True)   # drop NA
medical_join.drop_duplicates(inplace = True)   # drop duplicate rows
print(medical_join.shape[0])
medical_join.head()
```

```python
# save EMS15
data_ems15 = medical_join[['STARFIRE_INCIDENT_ID', 'timestamp', 'lat', 'lon']].copy()
data_ems15.to_csv('./data/ems15.csv')
```

```python
# fire
nonmed_join = pd.merge(data_nonmed, box_loc, on = 'ALARM_BOX_LOCATION')
print(nonmed_join.shape[0])
nonmed_join.dropna(inplace = True)   # drop NA
nonmed_join.drop_duplicates(inplace = True)   # drop duplicate rows
print(nonmed_join.shape[0])
nonmed_join.head()
```

```python
# categorize rescue/fire
print(nonmed_join['INCIDENT_CLASSIFICATION_GROUP'].unique())

def classify_fire(category):
    if category == 'NonMedical Emergencies' or category == 'NonMedical MFAs':
        return 'RESCUE'
    elif category == 'Structural Fires' or category == 'NonStructural Fires':
        return 'FIRE'
    else:
        raise Exception('Unknown fire category.')

nonmed_join['CATEGORY'] = nonmed_join['INCIDENT_CLASSIFICATION_GROUP'].apply(classify_fire)
nonmed_join.head()
```

```python
print('#rows rescue:', nonmed_join[nonmed_join['CATEGORY']=='RESCUE'].shape[0])
print('#rows fire:', nonmed_join[nonmed_join['CATEGORY']=='FIRE'].shape[0])
```

```python
# save fire15
data_fire15 = nonmed_join[['STARFIRE_INCIDENT_ID', 'timestamp', 'lat', 'lon', 'CATEGORY']].copy()
data_fire15.to_csv('./data/fire15.csv')
```
